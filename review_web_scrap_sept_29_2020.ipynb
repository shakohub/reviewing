```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
URL = 'https://simple.wikipedia.org/wiki/List_of_counties_in_Maryland'
```

```python
response= requests.get(URL)
response.status_code
```

```python
print(response.text[:300])
```

```python
raw_HTML = BeautifulSoup(response.text,'html.parser')
```

```python
raw_HTML.title
```

```python
raw_HTML.find('table')
```

```python
raw_HTML.find('table',class_='wikitable sortable')
```

```python
table = raw_HTML.find('table')
```

```python
table.find_all('tr')[:4]
```

```python
tabledata= table.find_all('tr')[1:]
```

```python
counties_info = {'FIPS_code':[],
                 'County_seat':[],
                 'Est.':[],
                 'Origin':[],
                 'Etymology':[],
                 'Population':[],
                 'Map':[],
                }
```

```python
counties_info.values()
```

```python
for index,item in enumerate(counties_info.values()):
    print(index,item)
```

```python
for raw in tabledata :
    raw_data = raw.find_all('td')
    
    for index, item in enumerate(counties_info.values()):
        
        data = raw_data[index].text
        
        if (data =="\n") or (data ==''):
            
            item.append(np.nan)
            
        else :
            item.append(data)
```

```python
counties_info['counties']
```

```python
counties_df = pd.DataFrame(data= counties_info)
```
